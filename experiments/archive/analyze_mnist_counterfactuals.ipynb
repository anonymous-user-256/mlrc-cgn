### Extract `ResNet18` features

```python
# model_type = "resnet18"
# model_type = "CNN trained on original dataset"
model_type = "CNN trained on counterfactual dataset"
```

```python
if model_type == "resnet18":
    model = models.resnet18(pretrained=True)
    model.fc = torch.nn.Identity()
elif model_type == "CNN trained on original dataset":
    ckpt_path = "../cgn_framework/mnists/experiments/classifier_colored_MNIST/weights/ckp_epoch_10.pth"
    model = CNN()
    model.load_state_dict(torch.load(ckpt_path, map_location="cpu"))
    model.cls = torch.nn.Identity()
elif model_type == "CNN trained on counterfactual dataset":
    ckpt_path = f"../cgn_framework/mnists/experiments/classifier_{dataset}_counterfactual/weights/ckp_epoch_10.pth"
    model = CNN()
    model.load_state_dict(torch.load(ckpt_path, map_location="cpu"))
    model.cls = torch.nn.Identity()
else:
    raise ValueError(f"Could not identify model type {model_type}")
```

```python
model = model.eval()
```

```python
def get_model_features(model, dl, num_batches_to_use=None):
    iterator = tqdm(
        dl,
        desc="Extracting features",
        bar_format='{l_bar}{bar:10}{r_bar}{bar:-10b}',
    )
    fvecs = []
    labels = []
    ctr = 0
    for (data, label) in iterator:
        fvec = model(data)
        fvecs.append(fvec)
        labels.append(label)
        
        if ctr == num_batches_to_use:
            break
        
        ctr += 1

    fvecs = torch.cat(fvecs, dim=0)
    labels = torch.cat(labels, dim=0)
    return fvecs, labels
```

```python
features_og, y_og = get_model_features(model, dl_og, num_batches_to_use=20)
```

```python
features_cf, y_cf = get_model_features(model, dl_cf, num_batches_to_use=20)
```

### Visualize TSNE

```python
def reduce_dimensionality(X, dim=2):
    assert len(X.shape) == 2
    N, D = X.shape
    assert D >= dim
    
    if isinstance(X, torch.Tensor):
        X = X.detach().cpu().numpy()

    tsne = TSNE(n_components=dim)
    Z = tsne.fit_transform(X)
    return Z
```

```python
Z_og = reduce_dimensionality(features_og)
```

```python
Z_cf = reduce_dimensionality(features_cf)
```

```python
Z_og.shape, Z_cf.shape
```

### Visualize scatter plot

```python
def create_df(Z, y):
    df = pd.DataFrame(None)
    df["Z1"] = Z[:, 0]
    df["Z2"] = Z[:, 1]
    df["y"] = y
    return df
```

```python
df_og = create_df(Z_og, y_og)
df_cf = create_df(Z_cf, y_cf)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 8), constrained_layout=True)

ax[0].grid()
ax[0].set_title("Original", fontsize=20)
sns.scatterplot(data=df_og, x="Z1", y="Z2", hue="y", palette="deep", ax=ax[0])

ax[1].grid()
ax[1].set_title("Counterfactual", fontsize=20)
sns.scatterplot(data=df_cf, x="Z1", y="Z2", hue="y", palette="deep", ax=ax[1])

plt.suptitle(f"Features for  {model_type} on {dataset.replace('_', ' ')}", fontsize=25)
save_path = join(REPO_PATH, f"experiments/results/feature_analysis_{model_type.replace(' ', '_')}_{dataset}.pdf")
plt.savefig(save_path, bbox_inches="tight", format="pdf")
plt.show()
```
